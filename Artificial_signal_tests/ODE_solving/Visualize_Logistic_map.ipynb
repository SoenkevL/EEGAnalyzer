```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
logMapData = pd.read_csv('Logistic_map.csv', header=None, index_col=None)
```

```python
logMapData.head()
```

```python
r_values = logMapData.iloc[:,0]
timeseries = logMapData.iloc[:,900:]
```

```python
r_values_map = np.ones(timeseries.shape)
for i, r in enumerate(r_values):
    r_values_map[i,:] = r
```

```python
plt.scatter(r_values_map.flatten(), timeseries.to_numpy().flatten())
```

```python
timeseries_round = np.round(timeseries, 3)
```

```python
plt.scatter(r_values_map.flatten(), timeseries_round.to_numpy().flatten())
```
